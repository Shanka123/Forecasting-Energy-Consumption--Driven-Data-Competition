```python
def test_ordinal(y,p,test_size):
    ord_list=[]
    count=0
    y=y.reset_index()
    del y['index']
    y['hour']=pd.to_datetime(y['Timestamp'],format='%Y-%m-%d %H:%M:%S').dt.hour.values.tolist()
    y['minute']=pd.to_datetime(y['Timestamp'],format='%Y-%m-%d %H:%M:%S').dt.minute.values.tolist()
    y['seconds']=3600*y['hour']+60*y['minute']
    y['ordinal'] = pd.to_datetime(y['Timestamp'],format='%Y-%m-%d %H:%M:%S').apply(lambda x: x.toordinal())
    
    y['ordinal']=y['ordinal']-p
    size=y.shape[0]
    if abs(y['minute'][size-1]-y['minute'][size-2])==15 or abs(y['minute'][size-1]-y['minute'][size-2])==45 or abs(y['hour'][size-1]-y['hour'][size-2])==5:
        m=int((86400-y['seconds'][size-1])/900)
        for i in range(m-1):
            ord_list.append(y['ordinal'][size-1])
            count=count+1
            if count==test_size:
                break
        k=y['ordinal'][size-1]
        while count<test_size:
            k=k+1
            for i in range(96):
                ord_list.append(k)
                count=count+1
                if count==test_size:
                    break
    
        
    elif abs(y['minute'][size-1]-y['minute'][size-2])==30:
        m=int((86400-y['seconds'][size-1])/1800)
        for i in range(m-1):
            ord_list.append(y['ordinal'][size-1])
            count=count+1
            if count==test_size:
                break
        k=y['ordinal'][size-1]
        while count<test_size:
            k=k+1
            for i in range(48):
                ord_list.append(k)
                count=count+1
                if count==test_size:
                    break
    
    
    elif abs(y['minute'][size-1]-y['minute'][size-2])==0 and  (abs(y['hour'][size-1]-y['hour'][size-2])==1 or 
                                                               abs(y['hour'][size-1]-y['hour'][size-2])==2 or
                                                               abs(y['hour'][size-1]-y['hour'][size-2])==20 or
                                                               abs(y['hour'][size-1]-y['hour'][size-2])==21 or
                                                               abs(y['hour'][size-1]-y['hour'][size-2])==23):
        m=int((86400-y['seconds'][size-1])/3600)
        for i in range(m-1):
            ord_list.append(y['ordinal'][size-1])
            count=count+1
            if count==test_size:
                break
        k=y['ordinal'][size-1]
        while count<test_size:
            k=k+1
            for i in range(24):
                ord_list.append(k)
                count=count+1
                if count==test_size:
                    break
    
        
    elif abs(y['minute'][size-1]-y['minute'][size-2])==0 and  abs(y['hour'][size-1]-y['hour'][size-2])==0:
        k=y['ordinal'][size-1]
        while count<test_size:
            k=k+1
            
            ord_list.append(k)
            count=count+1
    return ord_list
```
